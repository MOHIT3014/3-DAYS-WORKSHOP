```python
line_data =df.groupby('Month')['ArrDelay'].mean().reset_index()
```

```python
line_data
```

```python
fig =go.Figure(data = go.Scatter(x=line_data['Month'],y=line_data['ArrDelay'], mode='lines',marker=dict(color='green')))
fig.update_layout(title='Monthly Average Delay of Flights', xaxis_title ='Months', yaxis_title='Delayof Flight Arrival(in minutes)')
fig.show()
```
